## Linear Algebra
* Vector Operation
* Matrix Operation
* Eigen value and Eigen Vector
* SVD
* Low Rank Approximation




```python
import numpy as np
```

##Vector Declaration

```python
a=np.array([1,0,0])
b=np.array([0,1,0])
```

##Vector Operations


```python
print("vector A: ",a)
print("vector B: ",b)

print("Sum of vectors: ",a+b)
print("Diff of vectors: ",a-b)

d=np.dot(a,b)
print("Dot Product: ",d)

s1=np.linalg.norm(a)
print("Magnitude of vector A: ",s1)
s2=np.linalg.norm(b)
print("Magnitude of vector B: ",s2)

theta=np.arccos(d/(s1*s2))
print("Angle between vectors: ",np.degrees(theta))
```

## Matrix Declarations

```python
x=np.array([[1,0,0],[0,0,0],[0,0,1]])
y=np.array([[0,0,0],[0,1,0],[0,0,0]])
```

##Matrix Operations

```python
d=x+y
print("Matrix D: ",d)
print("Matrix D transpose: ",d.T)

dete=np.linalg.det(d)
print("Determinant: ",dete)


matmul=np.dot(x,y)
print("Matrix multiplication: ",matmul)


invq=np.linalg.inv(d)
print("Inverse: ",invq)

shape=np.shape(d)
print("Shape: ",shape)
```

##Gram Schmidt Orthogonalisation

```python
A = np.array([
    [2, 0],
    [0, 0],
    [0, 2]
], dtype=float)
a1 = A[:, 0]
q1 = a1 / np.linalg.norm(a1)
a2 = A[:, 1]
proj = np.dot(a2, q1) * q1
u2 = a2 - proj
q2 = u2 / np.linalg.norm(u2)
Q = np.column_stack((q1, q2))

print("Q matrix:")
print(Q)
```

## Eigenvalues and EigenVectors


```python
z=np.array([[2,-2,3],[1,1,1],[1,3,-1]])
eigenval,eigenvec=np.linalg.eig(z)
print("EigenValues: ",eigenval)
print("Eigenvector: ",eigenvec)
print("Determinant:" ,np.linalg.det(z))
```

##Singular Value Decomposition



```python
np.set_printoptions(precision=4,suppress=True)
```

```python
p=np.array([[1,2,3,-1,2],[0,1,3,2,-1],[1,1,-2,0,1],[2,1,3,-1,0]])
u,s,v=np.linalg.svd(p,full_matrices=True)
#u,s,v=np.linalg.svd(p,full_matrices=False) -> it will reduce matrix size
print("U: \n",u)
print("S: \n",s)
print("V: \n",v)
print("\n")
print(np.shape(u))
print(np.shape(v.T))
sd=np.zeros((4,5))
sd[:,:4]=np.diag(s)
print("\nMatrix Multiplication of u and u transpose:\n",np.dot(u,u.T))
```

##Low Rank Approximation

```python
c = 3
sdp = np.zeros((c, c))
sdp[:, :c] = sd[:c, :c]
pp=np.dot(u[:,:c],np.dot(sdp,v[:c,:]))
print("Low Rank Approximation: \n",pp)
print("Og matrix: \n",p)
```
