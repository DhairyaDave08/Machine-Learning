```python
import sympy as sym
```

```python
x=sym.symbols('x')
y=sym.sin(x*sym.log(x))
print(y)
```

```python
dy=sym.diff(y,x,1)
d2y=sym.diff(y,x,2)
print(dy)
print(d2y)
# General Structure: sym.diff(f(x),x,ith derivative)
```

```python
x=sym.symbols('x')
func=sym.exp(-3*x**2)
ans=sym.integrate(func,x)
print(ans)
ans_val=sym.integrate(func,(x,-sym.oo,sym.oo))
print(ans_val)

# General Structure: sym_integrate(f(x),x)
# Definite: sym_integrate(f(x),(x,lower_limit,upper_limit))
```

```python
x=sym.symbols('x')
z=sym.symbols('z')
func1=x**2+sym.exp(-z)
df_dx=sym.diff(func1,x)
df_dz=sym.diff(func1,z)
print(df_dx)
print(df_dz)


#Double Integration
i=sym.integrate(func1,(x,0,1),(z,0,1))
print(i)
```

```python
lim_x=sym.limit(func1,x,0)
lim_z=sym.limit(func1,z,sym.oo)
print(lim_x)
print(lim_z)
```

```python
##Taylor Series
func=sym.cos(x)
expand=sym.series(func,x,0,10)
print(expand)

#General Structure: sym.series(y,x,around_value, approximate n)
```
