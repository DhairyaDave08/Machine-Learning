## **Linear Regression Model**




```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Defining the Parameters
m=8
c=3
X=np.array([1,3,10,15,8,5,7,2,4,11,19,23,0,4,2,10])
Y=np.zeros(X.shape)
Y=m*X+c
err=np.random.normal(0,5,X.shape)
Y_err=Y+err
plt.scatter(X, Y,color='Blue')
plt.scatter(X, Y_err,color='Red')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Linear Graph")

plt.show()
```

```python
X_mean = np.mean(X)
Y_mean = np.mean(Y_err)

Sxx = np.sum((X - X_mean)**2)
Sxy = np.sum((X - X_mean)*(Y_err - Y_mean))

m_hat = Sxy / Sxx
c_hat = Y_mean - m_hat*X_mean

Y_pred = m_hat*X + c_hat

residuals = Y_err - Y_pred

sigma2_hat = np.sum(residuals**2)/(len(X)-2)

print("Actual m:", m)
print("Estimated m:", m_hat)

print("Actual c:", c)
print("Estimated c:", c_hat)

print("Estimated Variance:", sigma2_hat)

plt.scatter(X, Y_err, color='red', label='Noisy Data')
plt.plot(X, Y_pred, color='blue', label='Regression Line')

plt.xlabel("X")
plt.ylabel("Y")

plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(Y_err, Y_pred)
r2 = r2_score(Y_err, Y_pred)

print("MSE:", mse)
print("R² Score:", r2)
```
