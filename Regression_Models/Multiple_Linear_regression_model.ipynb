# **Multiple Linear Regresiion**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_regression
import plotly.graph_objects as go
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
X,Y=make_regression(n_samples=300,n_features=2,n_informative=2,n_targets=1,noise=35)
df=pd.DataFrame(X,columns=['x1','x2'])
df['y']=Y
df.head()
```

```python
# Ploting the data in 3D
fig = go.Figure(
    data=[
        go.Scatter3d(
            x=X[:, 0],      # Feature 1
            y=X[:, 1],      # Feature 2
            z=Y,            # Target
            mode='markers',
            marker=dict(
                size=4,
                color=Y,
                colorscale='Viridis',
                opacity=0.8
            )
        )
    ]
)

fig.update_layout(
    title='3D Regression Dataset',
    scene=dict(
        xaxis_title='X1',
        yaxis_title='X2',
        zaxis_title='Y'
    )
)

fig.show()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.25,random_state=42)
```

```python
# applying Regressions
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
lr.fit(X_train,Y_train)
```

```python
y_pred=lr.predict(X_test)
```

```python
# Calculating the metrics value
mae=mean_absolute_error(Y_test,y_pred)
mse=mean_squared_error(Y_test,y_pred)
r2=r2_score(Y_test,y_pred)
print('MAE:',mae)
print('MSE:',mse)
print('R2:',r2)
```

```python
w1 = lr.coef_[0]
w2 = lr.coef_[1]
b = lr.intercept_
```

```python
fig = go.Figure()

fig.add_trace(
    go.Scatter3d(
        x=X[:, 0],
        y=X[:, 1],
        z=Y,
        mode='markers',
        marker=dict(size=4),
        name='Data'
    )
)

x1 = np.linspace(X[:, 0].min(), X[:, 0].max(), 30)
x2 = np.linspace(X[:, 1].min(), X[:, 1].max(), 30)

x1g, x2g = np.meshgrid(x1, x2)
yg = lr.intercept_ + lr.coef_[0] * x1g + lr.coef_[1] * x2g
fig.add_trace(
    go.Surface(
        x=x1g,
        y=x2g,
        z=yg,
        opacity=0.6,
        name='Regression Plane'
    )
)

fig.update_layout(
    title='Multiple Linear Regression Plane',
    scene=dict(
        xaxis_title='X1',
        yaxis_title='X2',
        zaxis_title='Y'
    )
)

fig.show()
```
