```python
#Creating dataframes from lists
a=[[100,9.3,14],
   [80,8.9,12],
   [75,7.9,15],
   [100,9.1,20],
   [60,8,2]]
a_df=pd.DataFrame(a,columns=['Marks','CGPA','Package(in LPA)'])
a_df
```

```python
# Drop the Null Values
# in dataframe, it deletes the row if any coloumn in that row has Null value
print("The original size: ",df.shape) # originally it was 1629 rows
#df=df.dropna() # using this will lead to permanent change in the data #288
print("The new size: ",df.shape)
df=df.dropna(how="all") # drops that row which has all as Null value
df.shape #1629
df=df.dropna(subset=["poster_path","actors"])
df.shape #1524
```
