### **Attributes**

```python
#Attributes of the series
s=pd.Series([1,2,3,4],index=['A','B','C','D'],name="Sample")
print(s)
print("\n")

#Values in series
print(s.values)
print("\n")

#Indexes in Series
print("The index is: ",s.index)
print("\n")

#Datatype of series
print(s.dtype)
print("\n")

#Printing Single Element
print("The element in Bth index is: ",s['B'])
print("\n")


#Arithmetic Operations
s=s*8+13
s
```

```python
#Attributes

print(b.head()) #first 5 : default 5
print("\n")
print(b.tail(3)) # last 3 : default 5
print("\n")
print(b.sample(10)) #randomly selecting sample: default 1
print("\n")
print(b.value_counts()) # counts the frequency of the element
print("\n")
print(b.sort_values(ascending=False)) # deafult sorting is acending, here is descending # here lexicograpphical sorting is done
print("\n")
print(b.sort_index()) # index wise sorting will be done
```
