```python
import pandas as pd

# Define the dataframe for these plots within this cell to ensure it's correct
d = {
    "Age": [18, 20, 21, 19, 22, 23, 20],
    "Marks": [70, 85, 90, 75, 95, 88, 80],
    "Gender": ["M", "F", "M", "F", "M", "F", "M"]
}
df_student = pd.DataFrame(d)

sns.countplot(x="Gender", data=df_student)
plt.show()

sns.histplot(x="Marks", data=df_student)
plt.show()

sns.scatterplot(
    x="Age",
    y="Marks",
    data=df_student
)
plt.show()

sns.boxplot(x="Marks", data=df_student)
plt.show()
```
